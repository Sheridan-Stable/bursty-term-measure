##GENIA corpus

```python
json_file_path = data_path + 'GENIAcorpus3.02-preprocessed.json'

with open(json_file_path, "r") as j:
  genia = json.loads(j.read())
```

# Processing Data

```python
counter = CountVectorizer()
vect_col = counter.fit_transform(genia)
vocab = counter.get_feature_names_out()
```

## **Standard Text Analysis Variables and Functions**

```python
m = len(counter.get_feature_names_out())
d = vect_col.shape[0]
N_i = get_Ni(vect_col)
N_j = get_Nj(vect_col)
N = get_N(N_i)
B_ij = get_Bij(vect_col)
B_i = get_Bi(B_ij)
CF = get_cf(N_i)
DF = get_df(B_i, d)
IDF = get_idf(DF)
ICF = get_icf(CF)
```

# **Novel Text Anlalysis Functions**

```python
theta = get_thetas(CF, 1/1000)
E_ICF = get_eicf(theta, N)
uBi = get_meanBi(theta, N_j, d)
bernoulli = get_matrix_bernoulli(theta, d)
varianceBi = get_varianceBi(N_j, bernoulli)
E_IDF = get_eidf(uBi, varianceBi, d)
mean_Nj = N_j.mean()
constant = np.log(mean_Nj) + (-2 + mean_Nj)/(4*N*mean_Nj)
line = E_IDF + constant
```

# **Plots**

```python
X = IDF.A[0]
Y = ICF.A[0]
```

```python
# Here we remove all points in the plot that have the same coordinates

points = list(zip(X, Y))
no_rep_points = list(set(points))
x_and_y = np.array(list(zip(*no_rep_points)))
x = x_and_y[0]
y = x_and_y[1]
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, s = 3, c = 'k')
ax.plot(E_IDF, E_ICF, 'r', label = 'Expected relation Eq. (3)')
ax.plot(E_IDF, line, 'b--', label='Asymptotic relation of Eq. (4)')
ax.set_title('GENIA Term corpus')
ax.set_xlabel('IDF')
ax.set_ylabel('ICF')
ax.legend()
# The line below writes the file in the same folder as this notebook, uncomment to rewrite.
#fig.savefig('figure-2.pdf')
```
